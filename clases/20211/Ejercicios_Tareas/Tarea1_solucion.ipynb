# Punto 1:
Escriba una función que reciba como insumo una lista y extraiga el máximo de la misma. (Es decir, una función que haga el equivalente a _max_, por supuesto no puede usar esta función dentro de la suya)

````python
def maximo( lista ):  
    maximo = lista[ 0 ]  
    for a in lista:  
        if a > maximo:  
            maximo = a  
    return maximo 
````

````python
print(maximo([1, 2, -8, 0]))  
````

Punto 2:
Escriba una función llamada `palabra_mas_larga` que reciba una cadena de caracteres que retorne una tupla con la palabra más larga y su longitud.

E.g. 'Eso es queso' -> ('queso', 5)    La palabra más larga es queso y tiene 5 caracteres.

````python
def palabra_mas_larga(words):  
    lista=words.split(' ')
    tuples=[]
    for n in lista:  
        tuples.append((n, len(n)))  
    
    word_len=[]
    for i in tuples:
        word_len.append(i[1])
    index_max=word_len.index(max(word_len))
    return tuples[index_max]
````

````python
palabra_mas_larga('Eso es queso')
````
