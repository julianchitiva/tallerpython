4. ¿Cómo se distribuyo el tipo de pago? Realice un gráfico circular [(PIE)](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.pie.html) para mostrar sus resultados. Pueden buscar diferentes [paletas de colores](https://matplotlib.org/stable/tutorials/colors/colormaps.html).

```python
tipos_pagos=(pagos
             .payment_type
             .value_counts(
                 normalize=True
             )) # Tabulo de manera normalizada
tipos_pagos
```

```python
## Grafica circular
fig = plt.figure(figsize=(10,10))
fig.patch.set_facecolor('white') ### Color de fondo
cmap = plt.get_cmap("rainbow") ## Seleccion de paleta de colores
plt.pie(x = tipos_pagos, ## Vector a pintas
        labels = tipos_pagos.index, ## Etiquetas
        colors = cmap(np
                      .linspace(0,
                                1,
                                tipos_pagos.size)
                     ), ## Colores
        shadow = True, 
        startangle = 5,
        explode = (0.1,0,0,0,0.2),
        textprops = {'color':"k",
                     'fontsize':14},
        autopct = '%.1f')
plt.title('Proporcion de pagos\npor medio de pago', 
          fontsize=20)
plt.savefig("tipos_pago.png", bbox_inches="tight")
plt.show()
```

5. ¿Cuánto fue el precio en promedio por tipo de producto vendido? Realice un gráfico de [(Barras)](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html) para mostrar sus resultados. 

```python
ventas = (ventas
 .merge(productos, 
        on='product_id')) # Tengo que juntar ventas y productos

# Con groupby se puede estudiar el promedio por tipo de producto
promedio_por_producto = (ventas
 .groupby('product_category_name_english')['price']
 .mean())

# Ordenamos
(promedio_por_producto
 .sort_values(ascending=False, 
              inplace=True))
```

```python
## Forma 1 
fig = plt.figure(figsize = (18,8))
plt.bar(x = promedio_por_producto.index,
        height = promedio_por_producto)
plt.ylabel('Precio promedio',
           fontsize=20)
plt.xlabel('Categoría producto',
           fontsize=20)
plt.xticks(rotation = 90) 
plt.title('Gasto Promedio', 
          fontsize=30)
plt.show()
```

```python
# Forma 2
(promedio_por_producto
 .sort_values(ascending=False)
 .plot(kind='bar', 
       figsize=(18,8)))
plt.ylabel('Precio promedio',
           fontsize=20)
plt.xlabel('Categoría producto',
           fontsize=20)
plt.xticks(rotation = 90) 
plt.title('Gasto Promedio', 
          fontsize=30)
plt.show()
```

7. En un solo gráfico muestre la distribución por tipo de pago (el gráfico circular que hizo anteriormente) y la serie de tiempo del total dinero por tipo de pago. 
   
Para la serie de tiempo realice los siguientes pasos:
   - Construya un dataframe con pagos únicos, es decir, que no se repitan los tipos de pago por orden.
   - Junte las venats con las entregas para tener las fechas.
   - Construya un DataFrame que junte el id de la orden, la fecha y los tipos de pago. 
   - Construya la Fecha.
   - Construya la serie de tiempo para cada medio de pago. Puede usar `groupby` luego `pivot_table` o directamente `pivot_table`

```python
pagos_unicos = (pagos
                .groupby(['order_id', 'payment_type'])
                [['payment_value']]
                .sum()
                .reset_index())
```

```python
ventas=(ventas
        .merge(entregas,
               on=['order_id'])) ## Para tener las fechas 
```

```python
serie_pagos = (ventas
               .loc[:,
                    ['order_id',
                     'order_purchase_timestamp']]
               .merge(pagos_unicos, 
                      on=['order_id']))
```

```python
serie_pagos['Fecha'] = (pd
                        .to_datetime(
                            serie_pagos
                            .order_purchase_timestamp)
                        .dt.date
                       )
```

```python
serie_pagos = (serie_pagos
               .pivot_table(index='Fecha',
                            columns='payment_type',
                            values='payment_value',
                            aggfunc='sum'))
```

```python
serie_pagos.head()
```

```python
serie_pagos.plot()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1,
                               ncols=2,
                               figsize=(18,8))

fig.patch.set_facecolor('white') ### Color de fondo

## Grafico circular
cmap = plt.get_cmap("rainbow") ## Seleccion de paleta de colores
ax1.pie(x = tipos_pagos, ## Vector a pintas
        labels = tipos_pagos.index, ## Etiquetas
        colors = cmap(np
                      .linspace(0,
                                1,
                                tipos_pagos.size)
                     ), ## Colores
        shadow = True, 
        startangle = 5,
        explode = (0.1,0,0,0,0.2),
        textprops = {'color':"k",
                     'fontsize':14},
        autopct = '%.1f')
ax1.set_title('Proporcion de pagos\npor medio de pago', 
          fontsize=18)

## Serie de tiempo 
serie_pagos.plot(ax= ax2)
ax2.set_ylabel('Reales')
ax2.set_title('Evolucion por medio de pago',
              fontsize=18)

## Ajuste de la grafica
fig.tight_layout() 
plt.show()
```

14. Construya las variables dia_llegada y dia_llegada_est que representen el día en la semana que llegó y el día en la semana que debían llegar los envios.  

```python
ventas['dia_llegada']=(pd
                       .to_datetime(
                           ventas
                           .order_delivered_customer_date
                       ))
```

```python
ventas.dia_llegada.dt.day_name()
```

```python
ventas['dia_llegada']=(pd
                       .to_datetime(
                           ventas
                           .order_delivered_customer_date
                       )
                       .dt
                       .day_name())
ventas['dia_llegada_est']=(pd
                           .to_datetime(
                               ventas
                               .order_estimated_delivery_date
                           )
                           .dt
                           .day_name())

ventas['dia_llegada'] = pd.Categorical(
    ventas['dia_llegada'], 
    categories=['Monday','Tuesday', 'Wednesday', 
                'Thursday', 'Friday', 'Saturday', 
                'Sunday'])

ventas['dia_llegada_est'] = pd.Categorical(
    ventas['dia_llegada_est'], 
    categories=['Monday','Tuesday', 'Wednesday', 
                'Thursday', 'Friday', 'Saturday', 
                'Sunday'])
```
