4. ¿Cómo se distribuyo el tipo de pago? Realice un gráfico circular [(PIE)](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.pie.html) para mostrar sus resultados. Pueden buscar diferentes [paletas de colores](https://matplotlib.org/stable/tutorials/colors/colormaps.html).

```python
tipos_pago = pagos.payment_type.value_counts(normalize=True)
```

```python
fig = plt.figure(figsize=(8,8))
fig.patch.set_facecolor('white')
colores = plt.get_cmap('rainbow')
plt.pie(x = tipos_pago, 
       labels= tipos_pago.index,
       textprops = {'color':'k',
                   'fontsize':14},
       explode = (0.3, 0,0,0,0.1), 
       shadow=True,
       autopct = '%.0f',
       colors= colores(np.linspace(0,1,tipos_pago.size)))
plt.title('Distribución del medio de pago', fontsize=20)
plt.savefig('tipos_pago.png', bbox_inches = 'tight')
plt.show()
```

7. En un solo gráfico muestre la distribución por tipo de pago (el gráfico circular que hizo anteriormente) y la serie de tiempo del total dinero por tipo de pago. 
   
Para la serie de tiempo realice los siguientes pasos:
   - Construya un dataframe con pagos únicos, es decir, que no se repitan los tipos de pago por orden.
   - Junte las venats con las entregas para tener las fechas.
   - Construya un DataFrame que junte el id de la orden, la fecha y los tipos de pago. 
   - Construya la Fecha.
   - Construya la serie de tiempo para cada medio de pago. Puede usar `groupby` luego `pivot_table` o directamente `pivot_table`

```python
pagos_unicos = pagos.groupby(['order_id','payment_type'])[['payment_value']].sum().reset_index()
pagos_unicos.head()
```

```python
ventas = ventas.merge(entregas)
```

```python
pagos_fecha = ventas.loc[:,['order_id','order_purchase_timestamp']].merge(pagos_unicos)
```

```python
pagos_fecha['order_purchase_timestamp'] = pd.to_datetime(pagos_fecha.order_purchase_timestamp)
```

```python
pagos_fecha['order_purchase_timestamp']  = pagos_fecha.order_purchase_timestamp.dt.date
```

```python
pagos_fecha=pagos_fecha.pivot_table(index='order_purchase_timestamp',
                                    columns='payment_type', 
                                    values='payment_value', 
                                    aggfunc=sum)
```

```python
pagos_fecha.plot(figsize=(16,8))
```

**combinar los graficos**

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, 
                               ncols=2, 
                               figsize=(16,8))
colores = plt.get_cmap('rainbow')
ax2.pie(x = tipos_pago, 
       labels= tipos_pago.index,
       textprops = {'color':'k',
                   'fontsize':14},
       explode = (0.3, 0,0,0,0.1), 
       shadow=True,
       autopct = '%.0f',
       colors= colores(np.linspace(0,1,tipos_pago.size)))
ax2.set_title('Distribución del medio de pago', fontsize=20)

pagos_fecha.plot(ax = ax1)

plt.suptitle('Comportamiento de los Tipos de pago', fontsize=30)
plt.show()
```

10. Construya la serie de tiempo del volumen diario de compras. [(gráfico de líneas)](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) Explore que hacen las funciones `diff`, `shift` y  `pct_change`
